```python
#Add Dependencies
import pandas as pd
import numpy as np
```

```python
read_file = "Data/FAN_REG-2018-2020.csv"
zip_file = "Data/zip-county.csv"
county_file="Data/2019_County_data.csv"
write_file = "Data/FAN_REG-2018-2020-county.csv"
```

```python
# Import our input dataset
attendee_df = pd.read_csv(read_file, index_col=False, dtype=str)
```

```python
attendee_df
```

```python
attendee_df['ZIP'] = attendee_df['Zip'].str[:5]
```

```python
attendee_df.drop(attendee_df.columns[attendee_df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
```

```python
attendee_df
```

```python
# Import our zip dataset
zip_df = pd.read_csv(zip_file , low_memory=False, index_col=False, dtype=str)
```

```python
# Import our county dataset
county_df = pd.read_csv(county_file , low_memory=False, index_col=0, dtype=str)
```

```python
zip_df.head()
```

```python
zip_df.info()
```

```python
zip_df['ZIP'] = zip_df['ZIP'].astype('str')
```

```python
zip_df['COUNTY'] = zip_df['COUNTY'].astype('str')
```

```python
#Reduce each zip code to single County
zip_df=zip_df.drop_duplicates(subset=['ZIP'], keep='first')
```

```python
zip_df.info()
```

```python
#oscars_df.film.value_counts()
zip_df.ZIP.value_counts()
```

```python
county_drop = ['STATE', 'COUNTY', 'STNAME', "POPEST_MALE", 'MEDIAN_AGE_TOT', 'MEDIAN_AGE_MALE','MEDIAN_AGE_FEM' ]
```

```python
county_df_2 = county_df.drop(columns=['STATE', 'COUNTY', 'STNAME', "POPEST_MALE", 'POPEST_FEM', 'MEDIAN_AGE_TOT', 'MEDIAN_AGE_MALE','MEDIAN_AGE_FEM' ])
county_df_2
```

```python
combined_df = pd.merge(attendee_df, zip_df, how='left', left_on='ZIP', right_on='ZIP' )
```

```python
combined_df
```

```python
combined_county_df = pd.merge(combined_df, county_df_2, how='left', left_on='COUNTY', right_on='FIPS' )
```

```python
combined_county_df
```

```python
combined_county_df.rename({'CTYNAME': 'County Name', 'POPESTIMATE': 'Population'}, axis=1, inplace=True)
```

```python
combined_county_df
```

```python
#Export dataset to CSV
combined_county_df.to_csv(write_file)
```
